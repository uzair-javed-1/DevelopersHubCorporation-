```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn provides built-in datasets for practice
# Dataset source: https://github.com/mwaskom/seaborn-data
# You can explore more datasets using:
# sns.get_dataset_names()

# Load Iris dataset for this task
iris = sns.load_dataset('iris')

# Basic exploration
print("Shape:", iris.shape)
print("\nColumns:", iris.columns)
print("\nFirst 5 rows:\n", iris.head())

# Info and statistics
print("\nInfo:")
iris.info()

print("\nDescribe:\n", iris.describe())

# ---------------- VISUALIZATION ---------------- #

# Scatter Plot: relationship between sepal length and width
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title("Sepal Length vs Width")
plt.show()

# Histogram: distribution of features
iris.hist(figsize=(10,8))
plt.suptitle("Feature Distributions")
plt.show()

# Box Plot: check for outliers (numeric columns only)
plt.figure(figsize=(8,6))
sns.boxplot(data=iris.drop(columns=['species']))
plt.title("Box Plot for Outliers")
plt.show()
```

```python
## Observations:
- Setosa species is clearly separable from others.
- Some variations appear in sepal width and petal length.
- Boxplots show few outlier-like points but dataset is clean and balanced.
```
